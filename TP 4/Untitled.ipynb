```python
def invariantDistribution(x, y, a=10):
    """Definition of the invariant distribution."""
    
    log_result = - x**2 / a**2 - y**2 -1/4 * (x**2/a**2 - y**2)**2
    
    return log_result
```

```python
def sample_P_i(x, y, i, sigma=3, a=10):
    """Symetric Random Walk """
    
    # Sample U
    u = np.random.rand()
    
    # New x
    if i == 0:
        new_x = np.random.normal(x, sigma**2)
    if i == 1:
        new_y = np.random.normal(y, sigma**2)
        
    # Acceptation rate
    log_pi_x_y = invariantDistribution(x, y, a=a)
    
    if i == 0:
        log_pi_new_x_y = invariantDistribution(new_x, y, a=a)
    else:
        log_pi_new_x_y = invariantDistribution(x, new_y, a=a)
        
    log_alpha = min(0, log_pi_new_x_y - log_pi_x_y)
    
    # Acceptation
    if np.log(u) < log_alpha:
        
        if i == 0:
            x = new_x
        else:
            y = new_y
        
    return x, y, log_alpha
```

```python
def samplingP(x_0, y_0, N_iter=2000, a=10, sigma_1=3, sigma_2=3):
    """Sample from the chain with kernel P."""
    
    # Initialisation of the points
    x = x_0
    y = y_0
    
    # Saving array for the means of the acceptation rates
    mean_alpha_x = []
    mean_alpha_y = []
    
    # Saving array of all the acceptation rates
    alpha_x = []
    alpha_y = []
    
    for i in range(N_iter):
        
        # Sample u
        u = np.random.rand()
        
        if u < 1/2:
            x, y, log_alpha_x = sample_P_i(x, y, 0, sigma=sigma_1, a=a)
            alpha_x.append(np.exp(log_alpha_x))
        else:
            x, y, log_alpha_y = sample_P_i(x, y, 1, sigma=sigma_2, a=a)
            alpha_y.append(np.exp(log_alpha_y))
            
        mean_alpha_x.append(np.mean(np.array(alpha_x)[-50:]))
        mean_alpha_y.append(np.mean(np.array(alpha_y)[-50:]))
    
    return [x, y, mean_alpha_x, mean_alpha_y]        
```

```python
def AdaptativeMH(X, l, N_iter=2000, N_Batch=50):
    """Executes the Adaptative MH with Gibbs Sampler."""
    
    # Counter for x and y when the acceptation ratio is accepted
    counter_x = 0
    counter_y = 0
    
    # Saving array for the mean of the alpha
    alphas_x = []
    alphas_y = []
    
    # Saving array for the acceptation ratio
    mean_alpha_x = []
    mean_alpha_y = []
    
    # Saving array for all the x and y
    x_l = []
    y_l = []
    
    for n in range(N_iter):
        
        for j in range(2):
            
            # Initialisation of x and y
            x = X[0]
            y = X[1]
            
            # Dimension to update
            x_j = X[j]
            l_j = l[j]
            
            # Proposal
            new_x_j = np.random.normal(x_j, np.exp(l_j)**2)
            
            # Acceptation ratio
            if j == 0:
                log_pi_x_j = invariantDistribution(x_j, y)
                log_pi_new_x_j = invariantDistribution(new_x_j, y)
                log_alpha = min(0, log_pi_new_x_j - log_pi_x_j)
                alphas_x.append(np.exp(log_alpha))
            
            else:
                log_pi_x_j = invariantDistribution(x, x_j)
                log_pi_new_x_j = invariantDistribution(x, new_x_j)
                log_alpha = min(0, log_pi_new_x_j - log_pi_x_j)
                alphas_y.append(np.exp(log_alpha))
                
           # Acceptation
            u = np.random.rand()
            if np.log(u) < log_alpha:
                X[j] = new_x_j
            
                if j == 0:
                    counter_x += 1 
                else:
                    counter_y += 1
                    
        
        # Update of the variance
        if n%N_Batch == 0:

            # Number of batch done
            k = n//N_Batch
            
            if k != 0:
                delta_j = min(0.01, k**(-1/2))
            else:
                delta_j = 0.01

            if counter_x/N_Batch > 0.24:
                l[0] = l[0] + delta_j
            else:
                l[0] = l[0] - delta_j

            if counter_y/N_Batch > 0.24:
                l[1] = l[1] + delta_j
            else:
                l[1] = l[1] - delta_j
            
            # Reinitialisation of the counters
            counter_y = 0     
            counter_x = 0
        
        # Computation of the mean
        mean_alpha_x.append(np.mean(np.array(alphas_x[-N_Batch:])))
        mean_alpha_y.append(np.mean(np.array(alphas_y[-N_Batch:])))
        
        # Update the saving array for the samples
        x_l.append(X[0])
        y_l.append(X[1])
    
    return x_l, y_l, mean_alpha_x, mean_alpha_y
```

```python
x_l, y_l, alpha_x, alpha_y = AdaptativeMH([0.5, 0.5], l=np.zeros(2), N_iter=10000, N_Batch=50)
```

```python
plt.plot(np.linspace(1, 10000, len(alpha_x)), alpha_x, label="x")
plt.plot(np.linspace(1, 10000, len(alpha_y)), alpha_y, label="y")
plt.grid(True)
plt.legend()
plt.show();
```
